```python
import pandas as pd
```

```python
path1 = './data/HAR/'
path2 = './results/feature50/'

filename1 = 'data'
filename2 = 'cluster_results_f50_w5_c6'

df = pd.read_csv(path1 + filename1 +'.csv')
result = pd.read_csv(path2 + filename2 + '.csv', index_col=0)
```

```python
# 클러스터링 결과 index를 문자형[a,f]으로 변환
assigned_cluster_lst = []
for i in range(ord('a'), ord('f')+1): assigned_cluster_lst.append(chr(i))
result.index = assigned_cluster_lst
```

```python
# 데이터 구조 변경
idx_lst = []

for i in range(result.shape[0]):
    # 빈 문자열인 경우 skip
    if result.iloc[i, 0].strip('[]').strip() == '':
        continue
    # 각 행(클러스터)에 속한 sample들의 index를 list로 묶기
    int_lst = [int(x) for x in result.iloc[i, 0].strip('[]').split(' ') if x.strip() != '']
    idx_lst.append(int_lst)

# 각 샘플에 clustering 결과 매핑
df['pred'] = 0
for i in range(len(assigned_cluster_lst)): 
    df.loc[idx_lst[i], 'pred'] = assigned_cluster_lst[i]
```
